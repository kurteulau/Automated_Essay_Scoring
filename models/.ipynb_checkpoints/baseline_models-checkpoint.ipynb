# Define performance metric

```python
def mcrmse(df_true_labels, df_pred):
    """
    Calculate MCRMSE from pandas dataframes of true labels and predictions
    
    Args:
    - df_true_labels (df): df that includes true labels, don't need to remove
      extraneous columns
    - df_pred (df): df that includes ONLY predictions (cannot include
      extraneous columns). For our case, should have exactly 6 columns

    Returns:
    - MCRMSE (float): mean columnwise root mean squared error
    """
    
    rmse_list = []

    for col in df_pred.columns:
        y_true = df_true_labels[col]
        y_pred = df_pred[col]
        rmse_list.append(mean_squared_error(y_true, y_pred, squared=False))

    mcrmse = np.array(rmse_list).mean()

    return mcrmse
```

## 2. Bag-of-Words Linear Regression

```python
def make_lin_reg_preds(X_train, y_train, X_val, features, rubric_col):
    """
    Generate predictions from features for a single rubric col, like vocabulary

    Args:
    - X_train, y_train, X_val (df)
    - features (list): features to include in model (not 'full_text')
    - rubric_col: the rubric column you want to make predictions for

    Returns:
    - preds (array): predictions for that rubric column
    """

    X_train = X_train[features]
    y_train = y_train[rubric_col]
    X_val = X_val[features]

    reg = LinearRegression().fit(X_train, y_train)
    
    preds = reg.predict(X_val)

    print(f'X_train R^2 for {rubric_col}: {round(reg.score(X_train, y_train), 2)}')

    return preds
```

### Validation

```python
features = X_train.columns[1:]
df_val_pred = pd.DataFrame()

for rubric_col in RUBRIC_COLS:
    df_val_pred[rubric_col] = make_lin_reg_preds(X_train, y_train, X_val, features, rubric_col)

df_val_preds_rounded = df_val_pred.applymap(lambda number: round(number * 2) / 2)

display(df_val_pred)
```

```python
print('MCRMSE for rounded predictions:', mcrmse(y_val, df_val_preds_rounded))
print('MCRMSE for NOT rounded predictions:', mcrmse(y_val, df_val_pred))
```

## 3. Bag-of-Words XGBoost Regression


```python
# https://www.kaggle.com/code/riteshsinha/feedback-prize-data-exploration
def search_best_params(X, y, label, verbose=False):
    regressor=xgb.XGBRegressor(eval_metric='rmse',
                           booster='gbtree',
                           objective = 'reg:squarederror',
                           gamma=0,subsample=0.9,
                           colsample_bytree=1,
                           min_child_weight=1, 
                           n_jobs=4,
                           seed=273
                          )
    
    # Uncomment if want to actually do grid search (causes exponential increase
    # in runtime)
    # param_grid = {"max_depth":    [3, 4, 5, 8],
    #           "n_estimators": [100, 200, 500, 600, 700, 1000],
    #           "learning_rate": [0.001, 0.01, 0.015, .25]
    #          }
    
    param_grid = {"max_depth":    [5],
              "n_estimators": [500],
              "learning_rate": [0.01]
             }
             
    search = GridSearchCV(regressor, param_grid, cv=5).fit(X, y)
    if verbose:
        print(f"The best hyperparameters for {label} are ", search.best_params_)
    return(search)

def get_final_model(X, y, search):
    regressor=xgb.XGBRegressor(learning_rate = search.best_params_["learning_rate"],
                               n_estimators  = search.best_params_["n_estimators"],
                               max_depth     = search.best_params_["max_depth"],
                               eval_metric='rmse',
                               booster='gbtree',
                               objective = 'reg:squarederror',
                               gamma=0,subsample=0.9,
                               colsample_bytree=1,
                               min_child_weight=1, 
                               n_jobs=4,
                               seed=0
                              )
    regressor.fit(X, y)
    return(regressor)
```

```python
def make_pred_xgb(X_train, y_train, X_val, label, verbose=False):
    """
    Generates predictions for a specific rubric column using XGBoost.
    Features need to be in X_train and X_val already.

    Args:
    - X_train, y_train, X_val (pandas df): pre-processed df's with features
      and labels. Features should be 2nd column onwards. Validation sets need to
      have column labels matching essay rubric analytic measures
    - label (string): name of analytic measure

    Returns:
    - numpy array of predictions for given label
    """
    # Select training data
    X = X_train.iloc[:, 1:] # don't include text_id
    y = y_train[label]

    # Find best hyperparameters and fit model
    search_results = search_best_params(X, y, label, verbose=False)
    xgb_model = get_final_model(X, y, search_results)

    # Make predictions
    X = X_val.iloc[:, 1:] # don't include text_id
    predictions = xgb_model.predict(X)

    return xgb_model, pd.Series(predictions)
```

### XGBoost Feature Importances

```python
def plot_feature_importance(rubric_col):
    """"Plot feature importances from specified XGBoost model"""
        
    if rubric_col == 'cohesion':
        model = cohesion_model
    elif rubric_col == 'syntax':
        model = syntax_model
    elif rubric_col == 'vocabulary':
        model = vocab_model
    elif rubric_col == 'phraseology':
        model = phraseology_model
    elif rubric_col == 'grammar':
        model = grammar_model
    elif rubric_col == 'conventions':
        model = conventions_model
    
    fig, ax = plt.subplots(figsize=(12, 9))
    plot_importance(model, ax=ax, height=0.5, title = f'XGBoost Feature Importance: {rubric_col}')
```

```python
for rubric_col in RUBRIC_COLS:
    plot_feature_importance(rubric_col)
```

```python
# Create pandas dataframe of feature importances for all rubric cols
cohesion_importances = cohesion_model.get_booster().get_score()
syntax_importances = syntax_model.get_booster().get_score()
vocab_importances = vocab_model.get_booster().get_score()
phraseology_importances = phraseology_model.get_booster().get_score()
grammar_importances = grammar_model.get_booster().get_score()
conventions_importances = conventions_model.get_booster().get_score()

data = [cohesion_importances,
        syntax_importances,
        vocab_importances,
        phraseology_importances,
        grammar_importances,
        conventions_importances
        ]
 
importance_df = pd.DataFrame.from_records(data, index = RUBRIC_COLS)
importance_pivot_df = pd.pivot_table(importance_df, columns=importance_df.index)

importance_df
```

```python
# Sort by different column as you see fit
importance_pivot_df.sort_values(by='conventions', ascending=False)
```

```python
# Features that are in top 10 of all rubric cols
def get_top_feature_importances(rubric_col, top_n=10):
    """Gets top n feature importances for a rubric column,
    defaults to top 10"""
    return set(importance_pivot_df[rubric_col].sort_values(ascending=False).iloc[:top_n].index)

top_cohesion = get_top_feature_importances('cohesion', 10)
top_syntax = get_top_feature_importances('syntax', 10)
top_vocab = get_top_feature_importances('vocabulary', 10)
top_phraseology = get_top_feature_importances('phraseology', 10)
top_grammar = get_top_feature_importances('grammar', 10)
top_conventions = get_top_feature_importances('conventions', 10)

top_cohesion.intersection(top_syntax) \
            .intersection(top_vocab) \
            .intersection(top_phraseology) \
            .intersection(top_grammar) \
            .intersection(top_conventions)
```

```python
# Use rounded predictions to create confusion matrix

def show_confusion_matrix(true_labels_df, pred_labels_df, rubric_col):
    """
    Given true labels and predictions for each rubric column, show 
    a confusion matrix for that rubric domain.
    """
    labels = np.arange(1, 5.1, 0.5).astype(str)

    cm = confusion_matrix(true_labels_df[rubric_col].astype(str), pred_labels_df[rubric_col].astype(str), labels=labels)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot()
    plt.title(f'XGBoost Confusion Matrix: {rubric_col}\n')
    plt.show()

show_confusion_matrix(y_val, xgb_pred_df_rounded, 'vocabulary')
```
