```python
# Use competition train data as our entire dataset
# Will extract a train, val, and test set from here
df_train = pd.read_csv('data/raw/train.csv')
df_train
```

# Train-Validation-Test Split

<font color='red'>WARNING: `iterative_train_test_split` from `skmultilearn.model_selection` has no easy way to set random state so results are not deterministic.</font>

```python
# Using the raw data ONLY, and using only the train.csv (the only file with labels)

# Shuffle the df_train rows
df_train = shuffle(df_train, random_state = RANDOM_STATE)

# Separate input features from output labels
features = ['text_id','full_text', 'length', 'paragraphs']
output = ['cohesion',	'syntax',	'vocabulary',	'phraseology',	'grammar',	'conventions']

# in order to have the essay index "pass through" the matrix-based stratification function
# below, the index is added to X before converted to a matrix
df_train['essay_index'] = df_train.index
index_and_features = ['essay_index','text_id','full_text']

X = df_train.loc[:, index_and_features]
y = df_train.loc[:, output]

# Note: using scikit-multilearn's iterative_train_test_split feature for multilabel, iteratively stratified
# test-train split, BUT this function needs data in matrix form, not pandas datafram
X_mat = X.to_numpy()
y_mat = y.to_numpy()

# training data split off
# X_train, X_val, y_train, y_val = iterative_train_test_split(X, y, random_state=RANDOM_STATE, train_size=0.8)
X_experimental_mat, y_experimental_mat, X_test_mat, y_test_mat = iterative_train_test_split(X_mat, y_mat, test_size=0.2)

# validation data is split off (20% of original, 25% of post-test split)
# X_train, X_val, y_train, y_val = iterative_train_test_split(X, y, random_state=RANDOM_STATE, train_size=0.8)
X_train_mat, y_train_mat, X_val_mat, y_val_mat = iterative_train_test_split(X_experimental_mat, y_experimental_mat, test_size=0.25)

# X_train, X_val, y_train, y_val = iterative_train_test_split(X, y, random_state=RANDOM_STATE, train_size=0.8)
# X_train_mat, y_train_mat, X_val_mat, y_val_mat = iterative_train_test_split(X_mat, y_mat, test_size=0.2)

# Turn into Pandas DataFrames
X_train = pd.DataFrame(data=X_train_mat, columns=index_and_features)
X_val = pd.DataFrame(data=X_val_mat, columns=index_and_features)
X_test = pd.DataFrame(data=X_test_mat, columns=index_and_features)
y_train = pd.DataFrame(data=y_train_mat, columns=output)
y_val = pd.DataFrame(data=y_val_mat, columns=output)
y_test = pd.DataFrame(data=y_test_mat, columns=output)

# Get passed-through index values
X_train_index = X_train['essay_index']
X_val_index = X_val['essay_index']
X_test_index = X_test['essay_index']

# Reset the index
X_train = X_train.set_index(X_train_index)
X_val = X_val.set_index(X_val_index)
X_test = X_test.set_index(X_test_index)
y_train = y_train.set_index(X_train_index)
y_val = y_val.set_index(X_val_index)
y_test = y_test.set_index(X_test_index)

# Drop 'essay_index' as a column from input (already have as an index)
X_train = X_train.drop('essay_index', axis=1)
X_val = X_val.drop('essay_index', axis=1)
X_test = X_test.drop('essay_index', axis=1)

# Add text_id to y (output) since we may need this as a master key when tokenizing later
y_train['text_id'] = X_train['text_id']
y_val['text_id'] = X_val['text_id']
y_test['text_id'] = X_test['text_id']
# moves text_id to first column
cols = y_train.columns.tolist()
cols = cols[-1:]+cols[:-1]
y_train = y_train[cols]
y_val = y_val[cols]
y_test = y_test[cols]

print(f'X_train shape: {X_train.shape}, y_train shape: {y_train.shape}')
print(f'X_val shape: {X_val.shape}, y_val shape: {y_val.shape}')
print(f'X_test shape: {X_test.shape}, y_test shape: {y_test.shape}')
```

```python
cohesion_train=y_train['cohesion']
cohesion_val=y_val['cohesion']
cohesion_test=y_test['cohesion']

syntax_train=y_train['syntax']
syntax_val=y_val['syntax']
syntax_test=y_test['syntax']

vocab_train=y_train['vocabulary']
vocab_val=y_val['vocabulary']
vocab_test=y_test['vocabulary']

phraseology_train=y_train['phraseology']
phraseology_val=y_val['phraseology']
phraseology_test=y_test['phraseology']

grammar_train=y_train['grammar']
grammar_val=y_val['grammar']
grammar_test=y_test['grammar']

conventions_train=y_train['conventions']
conventions_val=y_val['conventions']
conventions_test=y_test['conventions']

figure, axis = plt.subplots(2,3, figsize=(20,10))

#display cohesion
axis[0,0].hist(cohesion_train, label='train', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[0,0].hist(cohesion_val, label='val', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[0,0].hist(cohesion_test, label='test', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[0,0].legend()
axis[0,0].set_title('Cohesion Score Distribution')

#display syntax
axis[0,1].hist(syntax_train, label='train', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[0,1].hist(syntax_val, label='val', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[0,1].hist(syntax_test, label='test', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[0,1].legend()
axis[0,1].set_title('Syntax Score Distribution')

#display vocab
axis[0,2].hist(vocab_train, label='train', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[0,2].hist(vocab_val, label='val', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[0,2].hist(vocab_test, label='test', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[0,2].legend()
axis[0,2].set_title('Vocabulary Score Distribution')

#display phraseology
axis[1,0].hist(phraseology_train, label='train', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[1,0].hist(phraseology_val, label='val', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[1,0].hist(phraseology_test, label='test', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[1,0].legend()
axis[1,0].set_title('Phraseology Score Distribution')

#display grammar
axis[1,1].hist(grammar_train, label='train', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[1,1].hist(grammar_val, label='val', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[1,1].hist(grammar_test, label='test', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[1,1].legend()
axis[1,1].set_title('Grammar Score Distribution')

#display conventions
axis[1,2].hist(conventions_train, label='train', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[1,2].hist(conventions_val, label='val', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[1,2].hist(conventions_test, label='test', alpha=0.5, bins = 9, range=(0.75,5.25))
axis[1,2].legend()
axis[1,2].set_title('Conventions Score Distribution')

plt.show()
```
